## Remove tests where there were no activities

```python
thresh=100
```

```python
results_chal[results_chal.columns[3:]]=results_chal[results_chal.columns[3:]].astype("float64")
```

```python
results_chal=results_chal[results_chal[results_chal.columns[4]]>thresh]
```

```python
results_chal.columns
```

```python
mean_results=results_chal.groupby(['Filename']).mean()
```

```python
mean_results
```

```python
mean_results.index
```

```python
mean_results=mean_results.reindex(["CA_Test1", "CA_Test2","CA_Test6","CA_Test7","CA_Test8","CA_Test9","CA_Test10",
                                   "CA_Test11","CA_Test12",
                                   'CA_Activite_libre_S4', 'CA_Activite_libre_S7','CA_Activite_libre_S8','CA_Activite_libre_S9', 
                      'CA_Activite_libre_S10','CA_Activite_libre_S11','CA_Activite_libre_S12',
                       'CA_Simulation_S8','CA_Simulation_S9', 
                      'CA_Simulation_S10','CA_Simulation_S11','CA_Simulation_S12'])
```

```python
mean_results.columns
```

```python
mean_result_HAI=mean_results.iloc[:9,:]
```

```python
mean_result_ActLibre=mean_results.iloc[9:16,:]
```

```python
mean_result_Simulation=mean_results.iloc[16:,:]
```

```python
mean_results['% Time both hand not moving']+mean_results['% Time both moving']+mean_results['% Time nonpref mov/ pref not mov']+mean_results['% Time pref mov/ nonpref not mov']
```

```python
mean_results['% Time non prefered hand moving']+mean_results['% Time prefered hand moving']-mean_results['% Time both moving']+mean_results['% Time both hand not moving']
```

```python
plt.figure()
plt.plot(mean_result_HAI['% Time both hand not moving'],color='r',label="both not moving")
plt.plot(mean_result_HAI['% Time both moving'],color='g',label="both moving")
plt.plot(mean_result_HAI['% Time nonpref mov/ pref not mov'],color='m',label="only nonpref moving")
plt.plot(mean_result_HAI['% Time pref mov/ nonpref not mov'],color='b',label="only pref moving")
plt.ylabel("% Time")
plt.legend()
```
