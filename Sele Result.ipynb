```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
Results_SELE="d:\\Users\\al-abiad\\Desktop\\BB-Bim\\SELE\\SL_alltests.xlsx"
results_sele=pd.read_excel(Results_SELE) 
results_sele
```

```python
thresh=100
results_sele[results_sele.columns[3:]]=results_sele[results_sele.columns[3:]].astype("float64")
results_sele=results_sele[results_sele[results_sele.columns[4]]>thresh]
mean_results=results_sele.groupby(['Filename']).mean()
mean_results
```

```python
mean_results=mean_results.reindex(["SL_Test1", "SL_Test2","SL_Test3","SL_Test4","SL_Test6","SL_Test7","SL_Test8",
                                   "SL_Test10","SL_Test11","SL_Test12",
                                   'SL_Activite_libre_S4', 'SL_Activite_libre_S7','SL_Activite_libre_S8','SL_Activite_libre_S9', 
                      'SL_Activite_libre_S10','SL_Activite_libre_S11','SL_Activite_libre_S12','SL_Activite_libre_S14',
                       'SL_Simulation_S7','SL_Simulation_S8', 
                      'SL_Simulation_S11','SL_Simulation_S12','SL_Simulation_S14'])
```

```python
mean_results
```

```python
mean_result_HAI=mean_results.iloc[:10,:]
```

```python
mean_result_ActLibre=mean_results.iloc[10:18,:]
```

```python
mean_result_Simulation=mean_results.iloc[18:,:]
```

```python
plt.figure()
plt.plot(mean_result_HAI['% Time both hand not moving'],color='r',label="both not moving")
plt.plot(mean_result_HAI['% Time both moving'],color='g',label="both moving")
plt.plot(mean_result_HAI['% Time nonpref mov/ pref not mov'],color='m',label="only nonpref moving")
plt.plot(mean_result_HAI['% Time pref mov/ nonpref not mov'],color='b',label="only pref moving")
plt.ylabel("% Time")
plt.legend()
```

```python

fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(mean_result_HAI['% Time prefered hand moving'],color='r',label="pref moving")
ax1.plot(mean_result_HAI['% Time non prefered hand moving'],color='g',label="non pref moving")
ax1.legend()
ax2.plot(mean_result_HAI['Use ratio (calculated as ratio of time)'],color='k',label="Use ratio time",linewidth=3, alpha=0.2)

ax1.set_xlabel('Weekly test')

ax1.set_ylabel('Time %', color='g')
ax2.set_ylabel('Ratio')

ax2.legend()
```
